```python
def to_txt(go_dict, path):
    path = "GO/" + path + ".txt"
    
    with open(path, "w") as f:
        for gene, go in go_dict.items():
            line = str(gene) + "\t" + " ".join(list(map(str, go))) + "\n"
            f.write(line)
```
